```python
#Convertir en dataframe
df= pd.read_csv("Detalle precios y productos fabricados 2022.csv")
df.head()
```

```python
#Verificamos informaci√≥n del DataFrame
df.info()
```

## **1. Filtro por objeto**

```python
filtro1=df[df["NOMBRE.1"] == "PROVEEDORA DE SEGURIDAD INDUSTRIAL DEL GOLFO"]
filtro1.head()
```

Se muestran solo las compras hechas por la Proveedora de Seguridad Industrial del Golfo. (se utiliza la columna "Nombre.1"

```python
filtro1.to_csv("Filtro1.csv")
files.download("Filtro1.csv")
```

## **2. Filtro por filas**

```python
filtro2= df.iloc[11:21,: ]
filtro2
```

Se filtraron las siguientes 10 filas consecutivas, es decir, de la 11 hasta la 20. 

```python
filtro2.to_csv("Filtro2.csv")
files.download("Filtro2.csv")
```

## **3. Filtro por columnas**

```python
filtro3=df.iloc[ : , [1,3,11,13]] 
filtro3
```

Muestra solo el costo y margen total de cada factura, junto con el nombre del cliente y la fecha en la que se realiz√≥ la compra. 

```python
filtro3.to_csv("Filtro3.csv")
files.download("Filtro3.csv")
```

## **4. Filtro de filas vs columnas**

```python
df1= pd.read_csv("Detalle precios y productos fabricados 2022.csv", index_col=0)
filtro4=df1.loc[["F37042","F37255","F38887"], ["SUBTOTAL_PARTIDA"]]
filtro4
```

Muestra el total de las facturas "F37042","F37255" y "F38887". 

```python
filtro4.to_csv("Filtro4.csv")
files.download("Filtro4.csv")
```

## **5. Filtro por cabecera**

```python
filtro5=df.head(50)
filtro5
```

Se filtraron los primeros 50 registros de facturas. 

```python
filtro5.to_csv("Filtro5.csv")
files.download("Filtro5.csv")
```

## **6. Filtro por comparaci√≥n**

```python
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > df["MARGEN_TOTAL_CALCULADO"].mean()]
filtro6
```

Filtro las facturas que tienen un margen de ganancia mayor al margen promedio de las facturas, utilizando la columna "MARGEN_TOTAL_CALCULADO". 

```python
filtro6.to_csv("Filtro6.csv")
files.download("Filtro6.csv")
```

## **7. Filtro and**

```python
filtro7=df[(df["CVE_ART"] == "TP0011") & (df["SUBTOTAL_PARTIDA"] > 1000)]
filtro7
```

Se muestran solo las facturas cuyo art√≠culo sea el correspondiente a la clave "TP0011" y que el total de la factura supere los mil pesos, utilizando las columnas "CVE_ART" y "SUBTOTAL_PARTIDA". 

```python
filtro7.to_csv("Filtro7.csv")
files.download("Filtro7.csv")
```

## **8. Filtro or**

```python
#Filtro o
filtro8=df[(df["NOMBRE"] == "LETICIA RAMIREZ HERNANDEZ") | (df["NOMBRE.1"] == "CARCOUSTICS INDUSTRIAL DE MEXICO")]
filtro8
```

Se muestra las ventas hechas por Leticia Ram√≠rez o las compras hechas por la empresa Carcoustics Industrial, utilizando las columnas nombre y nombre.1 respectivamente. 

```python
filtro8.to_csv("Filtro8.csv")
files.download("Filtro8.csv")
```

## **9. Filtro not**

```python
filtro9=df[~(df["NOMBRE"] == "DIEGO ARMANDO LEON DURAN") & ~(df["NOMBRE.1"] == "FIDEICOMISO F/1596")]
filtro9
```

Se filtraron todas las facturas que no tuvieran como vendedor a Diego Armando o como comprador a "FIDEICOMISO F/1596", utilizando las columnas Nombre y Nombre.1 respectivamente. 

```python
filtro9.to_csv("Filtro9.csv")
files.download("Filtro9.csv")
```
